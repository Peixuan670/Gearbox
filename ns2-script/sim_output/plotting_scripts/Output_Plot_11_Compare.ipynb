# This script plot the output of test 11_Compare

```python
%matplotlib inline

import numpy as np
import pandas as pd
from collections import OrderedDict
import matplotlib.pyplot as plt
```

### HCS

```python
th_0_HCS = pd.read_table('../11_Compare/outputs/HCS/throughput_0.txt', header=None)
th_0_HCS.columns = ['Timestamp', 'Throughput']

th_1_HCS = pd.read_table('../11_Compare/outputs/HCS/throughput_1.txt', header=None)
th_1_HCS.columns = ['Timestamp', 'Throughput']

th_2_HCS = pd.read_table('../11_Compare/outputs/HCS/throughput_2.txt', header=None)
th_2_HCS.columns = ['Timestamp', 'Throughput']

th_3_HCS = pd.read_table('../11_Compare/outputs/HCS/throughput_3.txt', header=None)
th_3_HCS.columns = ['Timestamp', 'Throughput']
```

```python
th_0_HCS.head()
```

```python
print("th0_HCS.shape = ", th_0_HCS.shape)
print("th1_HCS.shape = ", th_1_HCS.shape)
print("th2_HCS.shape = ", th_2_HCS.shape)
print("th3_HCS.shape = ", th_3_HCS.shape)
```

```python
plt.figure(figsize=(15,8))
plt.xlabel("Time(sec)")
plt.ylabel("Throughput(bps)")
plt.title("Throughput of HCS")
plt.legend()
line = plt.axhline(y=1000, xmin=0, xmax=12, linewidth=2, color = 'violet')
line.set_dashes((10, 4))
#type(th_0_HCS)
plt.plot(th_0_HCS['Timestamp'], th_0_HCS['Throughput'], color='red', label='th0_HCS')
plt.plot(th_1_HCS['Timestamp'], th_1_HCS['Throughput'], color='blue', label='th1_HCS')
plt.plot(th_2_HCS['Timestamp'], th_2_HCS['Throughput'], color='green', label='th2_HCS')
plt.plot(th_3_HCS['Timestamp'], th_3_HCS['Throughput'], color='orange', label='th3_HCS')
```
