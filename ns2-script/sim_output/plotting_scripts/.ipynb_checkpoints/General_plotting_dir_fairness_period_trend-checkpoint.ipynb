### Fairness Measurement Function

```python
# Function to measure fairness

def measure_fairness(raw_data_0, raw_data_1, raw_data_2, raw_data_3, period, step):
    raw_data_0['Short_th'] = raw_data_0['Data'].shift(step)
    raw_data_1['Short_th'] = raw_data_1['Data'].shift(step)
    raw_data_2['Short_th'] = raw_data_2['Data'].shift(step)
    raw_data_3['Short_th'] = raw_data_3['Data'].shift(step)

    period_0 = raw_data_0.fillna(0)
    period_1 = raw_data_1.fillna(0)
    period_2 = raw_data_2.fillna(0)
    period_3 = raw_data_3.fillna(0)

    period_0['Short_th'] = period_0['Data'] - period_0['Short_th']
    period_1['Short_th'] = period_1['Data'] - period_1['Short_th']
    period_2['Short_th'] = period_2['Data'] - period_2['Short_th']
    period_3['Short_th'] = period_3['Data'] - period_3['Short_th']

    period_0['Short_th'] = period_0['Short_th']/period
    period_1['Short_th'] = period_1['Short_th']/period
    period_2['Short_th'] = period_2['Short_th']/period
    period_3['Short_th'] = period_3['Short_th']/period


    
    plt.figure(figsize=(8,5))
    plt.xlabel("Time(sec)")
    plt.ylabel("Throughput(bps)")
    plt.title("Short-tern Throughput")
    plt.legend()
    line = plt.axhline(y=1000, xmin=0, xmax=12, linewidth=2, color = 'violet')
    line.set_dashes((10, 4))
    #plt.xlim(0, 1)
    plt.ylim(-50, 1600)
    plt.plot(period_0['Timestamp'], period_0['Short_th'], color='red', label='th0')
    plt.plot(period_1['Timestamp'], period_1['Short_th'], color='blue', label='th1')
    plt.plot(period_2['Timestamp'], period_2['Short_th'], color='green', label='th2')
    plt.plot(period_3['Timestamp'], period_3['Short_th'], color='orange', label='th3')
    
    plt.figure(figsize=(8,5))
    plt.xlabel("Time(sec)")
    plt.ylabel("Throughput(bps)")
    plt.title("Short-tern Throughput of Gearbox")
    plt.legend()
    line = plt.axhline(y=1000, xmin=0, xmax=12, linewidth=2, color = 'violet')
    line.set_dashes((10, 4))
    #plt.xlim(0, 1)
    plt.ylim(-50, 1600)
    plt.plot(period_0['Timestamp'], period_0['Short_th'], color='red', label='th0_HCS')
    plt.plot(period_1['Timestamp'], period_1['Short_th'], color='blue', label='th1_HCS')
    plt.plot(period_2['Timestamp'], period_2['Short_th'], color='green', label='th2_HCS')
    plt.plot(period_3['Timestamp'], period_3['Short_th'], color='orange', label='th3_HCS')
```

### HCS

```python
# Loop
period_list = [0.01, 0.05, 0.1, 0.5, 1]
step_list = [1, 5, 10, 50, 100]

for i = np.linspace(0, len(period_list)):
    measure_fairness(raw_data_0_HCS, raw_data_1_HCS, raw_data_2_HCS, raw_data_3_HCS, period[i], step[i])
#period_list[0]
```

```python
# Short term fairness

period = 0.1
step = 10

raw_data_0_HCS['Short_th'] = raw_data_0_HCS['Data'].shift(step)
raw_data_1_HCS['Short_th'] = raw_data_1_HCS['Data'].shift(step)
raw_data_2_HCS['Short_th'] = raw_data_2_HCS['Data'].shift(step)
raw_data_3_HCS['Short_th'] = raw_data_3_HCS['Data'].shift(step)

period_0_HCS = raw_data_0_HCS.fillna(0)
period_1_HCS = raw_data_1_HCS.fillna(0)
period_2_HCS = raw_data_2_HCS.fillna(0)
period_3_HCS = raw_data_3_HCS.fillna(0)

period_0_HCS['Short_th'] = period_0_HCS['Data'] - period_0_HCS['Short_th']
period_1_HCS['Short_th'] = period_1_HCS['Data'] - period_1_HCS['Short_th']
period_2_HCS['Short_th'] = period_2_HCS['Data'] - period_2_HCS['Short_th']
period_3_HCS['Short_th'] = period_3_HCS['Data'] - period_3_HCS['Short_th']

period_0_HCS['Short_th'] = period_0_HCS['Short_th']/period
period_1_HCS['Short_th'] = period_1_HCS['Short_th']/period
period_2_HCS['Short_th'] = period_2_HCS['Short_th']/period
period_3_HCS['Short_th'] = period_3_HCS['Short_th']/period

#period_0_HCS.head(30)
```

```python
plt.figure(figsize=(8,5))
plt.xlabel("Time(sec)")
plt.ylabel("Throughput(bps)")
plt.title("Short-tern Throughput of Gearbox")
plt.legend()
line = plt.axhline(y=1000, xmin=0, xmax=12, linewidth=2, color = 'violet')
line.set_dashes((10, 4))
#plt.xlim(0, 1)
plt.ylim(-50, 1600)
plt.plot(period_0_HCS['Timestamp'], period_0_HCS['Short_th'], color='red', label='th0_HCS')
plt.plot(period_1_HCS['Timestamp'], period_1_HCS['Short_th'], color='blue', label='th1_HCS')
plt.plot(period_2_HCS['Timestamp'], period_2_HCS['Short_th'], color='green', label='th2_HCS')
plt.plot(period_3_HCS['Timestamp'], period_3_HCS['Short_th'], color='orange', label='th3_HCS')
```

```python
# Find max difference

expected_th = 1000

merge_short_fairness_HCS = pd.DataFrame()

merge_short_fairness_HCS['Timestamp'] = period_0_HCS['Timestamp']
merge_short_fairness_HCS['Short_th_flow0'] = period_0_HCS['Short_th']
merge_short_fairness_HCS['Short_th_flow1'] = period_1_HCS['Short_th']
merge_short_fairness_HCS['Short_th_flow2'] = period_2_HCS['Short_th']
merge_short_fairness_HCS['Short_th_flow3'] = period_3_HCS['Short_th']

merge_short_fairness_HCS['Max'] = merge_short_fairness_HCS.apply(lambda x: max(x['Short_th_flow0'],x['Short_th_flow1'],x['Short_th_flow2'],x['Short_th_flow3']), axis=1)
merge_short_fairness_HCS['Min'] = merge_short_fairness_HCS.apply(lambda x: min(x['Short_th_flow0'],x['Short_th_flow1'],x['Short_th_flow2'],x['Short_th_flow3']), axis=1)

merge_short_fairness_HCS['Short_fairness'] = merge_short_fairness_HCS.apply(lambda x: (x['Max']-x['Min'])/expected_th, axis=1)
merge_short_fairness_HCS.head()

plt.figure(figsize=(8,5))
plt.xlabel("Time(sec)")
plt.ylabel("Short-term Fairness")
plt.title("Fairness of AFQ-1000")
plt.legend()
plt.plot(merge_short_fairness_HCS['Timestamp'], merge_short_fairness_HCS['Short_fairness'], color='blue', label='fairness_AFQ-100')
```
