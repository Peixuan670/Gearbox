```python
# Short term fairness

period = 0.1
step = 10

raw_data_0_HCS['Short_th'] = raw_data_0_HCS['Data'].shift(step)
raw_data_1_HCS['Short_th'] = raw_data_1_HCS['Data'].shift(step)
raw_data_2_HCS['Short_th'] = raw_data_2_HCS['Data'].shift(step)
raw_data_3_HCS['Short_th'] = raw_data_3_HCS['Data'].shift(step)

period_0_HCS = raw_data_0_HCS.fillna(0)
period_1_HCS = raw_data_1_HCS.fillna(0)
period_2_HCS = raw_data_2_HCS.fillna(0)
period_3_HCS = raw_data_3_HCS.fillna(0)

period_0_HCS['Short_th'] = period_0_HCS['Data'] - period_0_HCS['Short_th']
period_1_HCS['Short_th'] = period_1_HCS['Data'] - period_1_HCS['Short_th']
period_2_HCS['Short_th'] = period_2_HCS['Data'] - period_2_HCS['Short_th']
period_3_HCS['Short_th'] = period_3_HCS['Data'] - period_3_HCS['Short_th']

period_0_HCS['Short_th'] = period_0_HCS['Short_th']/period
period_1_HCS['Short_th'] = period_1_HCS['Short_th']/period
period_2_HCS['Short_th'] = period_2_HCS['Short_th']/period
period_3_HCS['Short_th'] = period_3_HCS['Short_th']/period

#period_0_HCS.head(30)
```

```python
plt.figure(figsize=(8,5))
plt.xlabel("Time(sec)")
plt.ylabel("Throughput(bps)")
plt.title("Short-tern Throughput of Gearbox")
plt.legend()
line = plt.axhline(y=1000, xmin=0, xmax=12, linewidth=2, color = 'violet')
line.set_dashes((10, 4))
#plt.xlim(0, 1)
plt.ylim(-50, 1600)
plt.plot(period_0_HCS['Timestamp'], period_0_HCS['Short_th'], color='red', label='th0_HCS')
plt.plot(period_1_HCS['Timestamp'], period_1_HCS['Short_th'], color='blue', label='th1_HCS')
plt.plot(period_2_HCS['Timestamp'], period_2_HCS['Short_th'], color='green', label='th2_HCS')
plt.plot(period_3_HCS['Timestamp'], period_3_HCS['Short_th'], color='orange', label='th3_HCS')
```

```python
short_fairness_HCS = pd.merge(period_3_HCS, period_0_HCS, on='Timestamp', how='outer')
short_fairness_HCS['Fairness'] = (short_fairness_HCS['Short_th_x'] - short_fairness_HCS['Short_th_y'])/1000


plt.figure(figsize=(8,5))
plt.xlabel("Time(sec)")
plt.ylabel("Short-term Fairness")
plt.title("Fairness of Gearbox")
plt.legend()
#line = plt.axhline(y=1000, xmin=0, xmax=12, linewidth=2, color = 'violet')
#line.set_dashes((10, 4))
#plt.xlim(0, 1)
#plt.ylim(-50, 1600)
plt.plot(short_fairness_HCS['Timestamp'], short_fairness_HCS['Fairness'], color='blue', label='fairness_HCS')
```
