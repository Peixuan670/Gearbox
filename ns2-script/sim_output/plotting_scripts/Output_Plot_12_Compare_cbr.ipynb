# This script plot the output of test 12_Compare_cbr

```python
%matplotlib inline

import numpy as np
import pandas as pd
from collections import OrderedDict
import matplotlib.pyplot as plt
```

### HCS + 1Mbps CBR

```python
th_0_HCS_1m = pd.read_table('../12_Compare_cbr/outputs/HCS_1m/throughput_0.txt', header=None)
th_0_HCS_1m.columns = ['Timestamp', 'Throughput']

th_1_HCS_1m = pd.read_table('../12_Compare_cbr/outputs/HCS_1m/throughput_1.txt', header=None)
th_1_HCS_1m.columns = ['Timestamp', 'Throughput']

th_2_HCS_1m = pd.read_table('../12_Compare_cbr/outputs/HCS_1m/throughput_2.txt', header=None)
th_2_HCS_1m.columns = ['Timestamp', 'Throughput']

th_3_HCS_1m = pd.read_table('../12_Compare_cbr/outputs/HCS_1m/throughput_3.txt', header=None)
th_3_HCS_1m.columns = ['Timestamp', 'Throughput']
```

```python
th_0_HCS_1m.head()
```

```python
print("th0_HCS_1m.shape = ", th_0_HCS_1m.shape)
print("th1_HCS_1m.shape = ", th_1_HCS_1m.shape)
print("th2_HCS_1m.shape = ", th_2_HCS_1m.shape)
print("th3_HCS_1m.shape = ", th_3_HCS_1m.shape)
```

```python
plt.figure(figsize=(15,8))
plt.xlabel("Time(sec)")
plt.ylabel("Throughput(bps)")
plt.title("Throughput of HCS_1m")
plt.legend()
line = plt.axhline(y=1000, xmin=0, xmax=12, linewidth=2, color = 'violet')
line.set_dashes((10, 4))
#type(th_0_HCS)
plt.plot(th_0_HCS_1m['Timestamp'], th_0_HCS_1m['Throughput'], color='red', label='th0_HCS_1m')
plt.plot(th_1_HCS_1m['Timestamp'], th_1_HCS_1m['Throughput'], color='blue', label='th1_HCS_1m')
plt.plot(th_2_HCS_1m['Timestamp'], th_2_HCS_1m['Throughput'], color='green', label='th2_HCS_1m')
plt.plot(th_3_HCS_1m['Timestamp'], th_3_HCS_1m['Throughput'], color='orange', label='th3_HCS_1m')
```
