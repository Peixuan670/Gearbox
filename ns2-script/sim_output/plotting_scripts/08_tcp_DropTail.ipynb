### TCP Same Link Capacity + DropTail (All 4x2Mbps, bottleneck 4Mbps)

```python
th_0_same_DropTail = pd.read_table('../08_incast_4M_12s_tcp_DropTail/outputs/same_cap_DropTail/throughput_0.txt', header=None)
th_0_same_DropTail.columns = ['Timestamp', 'Throughput']

th_1_same_DropTail = pd.read_table('../08_incast_4M_12s_tcp_DropTail/outputs/same_cap_DropTail/throughput_1.txt', header=None)
th_1_same_DropTail.columns = ['Timestamp', 'Throughput']

th_2_same_DropTail = pd.read_table('../08_incast_4M_12s_tcp_DropTail/outputs/same_cap_DropTail/throughput_2.txt', header=None)
th_2_same_DropTail.columns = ['Timestamp', 'Throughput']

th_3_same_DropTail = pd.read_table('../08_incast_4M_12s_tcp_DropTail/outputs/same_cap_DropTail/throughput_3.txt', header=None)
th_3_same_DropTail.columns = ['Timestamp', 'Throughput']
```
